```python
import re
import pickle
from janome.tokenizer import Tokenizer

with open('animetitle_from_1970.txt', mode='r',encoding='utf-8') as f:
    anime = f.read()
# print(anime)
anime = re.sub('[() ❤★ 「」［］♪ ☆◎<>…｢｣【】]',"",anime)
anime = re.sub("みんなのうた","",anime)
# print(anime)
anime_list = anime.split('\n')
anime_list.pop()
# print(anime_list)
t = Tokenizer()

anime_words = []
for line in anime_list:
    anime_words.append(t.tokenize(line,wakati=True))
# print(anime_words)
with open('animetitle_from1970.pickle',mode='wb') as f:
    pickle.dump(anime_words,f)
```

```python
with open('animetitle_from1970.pickle',mode='rb') as f:
    anime_title = pickle.load(f)
print(anime_title)
```

```python
from gensim.models.doc2vec import Doc2Vec,TaggedDocument

tagged_documents = []
for i,line in enumerate(anime_title):
    tagged_documents.append(TaggedDocument(line,[i]))

model = Doc2Vec(documents=tagged_documents,
               vector_size = 100,
               min_count = 5,
               window=5,
               epochs=20,
               dm=0)
```

```python
print(anime_title[0])
print(model.docvecs[0])
print(model.docvecs[0].shape)
```

```python
len(anime_title)
```

```python
num = len(anime_title) -24
```

```python
print(model.docvecs.most_similar(num))
```

```python
print(anime_title[num])
print("--------------------------")
for p in model.docvecs.most_similar(num):
    print(anime_title[p[0]])
```
